```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("social.csv")
```

```python
df.head()
```

```python
features=df.iloc[:,:-1]
features
```

```python
label=df.iloc[:,-1]
label.head()
```

```python
actual_features=features.iloc[:,2:4]
actual_features
```

```python
actual_features.shape #feature scalling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
f_sc=StandardScaler()
#new training and testing data
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_ans,test_ans=train_test_split(actual_features,label,test_size=0.2)
```

```python
#here appy feature scaling apply
final_train=f_sc.fit_transform(train_data) #always max training data
```

```python
#testing data
final_test=f_sc.transform(test_data) #only testing data
```

```python
#calling random forest classifer
clf=RandomForestClassifier(n_estimators=30,criterion='entropy')
```

```python
trained=clf.fit(final_train,train_ans)
```

```python
predicted=trained.predict(final_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test_ans,predicted)
```
